```python
'''Librerias'''
import cv2
import os
import numpy as np
import glob
from sklearn.preprocessing import MinMaxScaler
import h5py
```

```python
'''Clases de datos'''
images_path = "C:/Users/Junior/Desktop/Proyecto mineria/Dataset_Aum_Black/"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
'''Caracteristicas del histograma'''
def fd_histogram(image, mask=None):
    #  Convertir a HSV color-space
    image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    # Computacinal a color histogram
    hist  = cv2.calcHist([image], [0, 1, 2], None, [bins, bins, bins], [0, 256, 0, 256, 0, 256])
    # Normalizar el histogram
    cv2.normalize(hist, hist)
    # Retorna el histogram
    return hist.flatten()

'''Determinar labels'''
color_features = []
color_features_labels = []
```

```python
'''Procesar las caracteristicas'''
print("[STATUS] Started extracting color features..")
for class_name in class_names:
    cur_path = images_path + "/" + class_name
    cur_label = class_name
    i = 1
    #Circulo de seleccionamiento de imagen
    for file in glob.glob(cur_path + "/*.png"):
        print("Processing Image - {} in {}".format(i, cur_label))
        # lectura de imagenes
        image = cv2.imread(file)

        # Extraer las caracteristicas
        bins = 8
        hist = fd_histogram(image)

        # Abrir las caracteristicas de histograma
        color_features.append(hist)
        color_features_labels.append(cur_label)

        # Mostrar la carga
        i += 1
```

```python
# get the overall feature vector size
print("[STATUS] color feature vector size {}".format(np.array(color_features).shape))
```

```python
# get the overall training label size
print("[STATUS] training Labels {}".format(np.array(color_features_labels).shape))
```

```python
# encode the target labels
def replace(my_list, my_dict):
    return [x if x not in my_dict else my_dict[x] for x in my_list]
labels_dict = {'car': 0, 'motor': 1}
target = replace(color_features_labels, labels_dict)

print("[STATUS] training labels encoded...")
```

```python
# scale features in the range (0-1)
scaler            = MinMaxScaler(feature_range=(0, 1))
rescaled_features = scaler.fit_transform(color_features)
print("[STATUS] feature vector normalized...")
print("[STATUS] target labels: {}".format(np.array(target)))
print("[STATUS] target labels shape: {}".format(np.array(target).shape))
```

```python
'''Guardar catacteristicas de vector using HDF5'''
h5f_data = h5py.File('C:/Users/Junior/Desktop/Proyecto mineria/Features_Data/color_features.h5', 'w')
h5f_data.create_dataset('Vehicle_Color_Features', data=np.array(rescaled_features))

'''Guardar label'''
h5f_label = h5py.File('C:/Users/Junior/Desktop/Proyecto mineria/Features_Data/color_features_labels.h5', 'w')
h5f_label.create_dataset('Vehicle_Color_Features_Labels', data=np.array(target))

'''Cerrar los datos'''
h5f_data.close()
h5f_label.close()

'''Imprimir el estado'''
print("[STATUS] end of training..")
```
